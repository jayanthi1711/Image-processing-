# morphological transformation


```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

# DILATION

```python
img=cv.imread('C:/Users/SN61/Downloads/dilation.png',0)
```

```python
img
```

```python
img=cv.imread('C:/Users/SN61/Downloads/dilation.png',0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,225,cv.THRESH_BINARY)
kernal=np.ones((3,3),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
titles=['image','mask','dilate']
images=[img,mask,dilation]
for i in range(len(images)):
    plt.subplot(1,3,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

# EROSION

```python
img=cv.imread('C:/Users/SN61/Downloads/dilation.png',0)
img=cv.cvtColor(img,cv.COLOR_BGR2RGB)
_,mask=cv.threshold(img,220,225,cv.THRESH_BINARY)
kernal=np.ones((3,3),np.uint8)
dilation=cv.dilate(mask,kernal,iterations=1)
erode=cv.erode(mask,kernal,iterations=1)
titles=['image','mask','dilate','erode']
images=[img,mask,dilation,erode]
for i in range(len(images)):
    plt.subplot(1,4,i+1),plt.imshow(images[i])
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

# MORPHOLOGICAL OPENING

```python
open =cv.morphologyEx(img,cv.MORPH_OPEN,kernal)
```

```python
cv.imshow('img',img)
cv.imshow('MORPH OPEN',open)
cv.waitKey(0)
cv.destroyAllWindows()
```

# MORPHOLOGICAL CLOSING

```python
close =cv.morphologyEx(img,cv.MORPH_CLOSE,kernal)
```

```python
cv.imshow('img',img)
cv.imshow('MORPH CLOSE',close)
cv.waitKey(0)
cv.destroyAllWindows()
```

# MORPHOLOGICAL GRADIENT-diff.of both dilation and errosion

```python
gradient=cv.morphologyEx(img,cv.MORPH_GRADIENT,kernal)
```

# display-morph gradient

```python
cv.imshow('img',img)
cv.imshow('MORPH GRADIENT',gradient)
cv.waitKey(0)
cv.destroyAllWindows()
```

# TOP HAT-opening & diff.of original & opening

```python
tophat=cv.morphologyEx(img,cv.MORPH_TOPHAT,kernal)
```

# display tophat

```python
cv.imshow('img',img)
cv.imshow('TOP HAT',tophat)
cv.waitKey(0)
cv.destroyAllWindows()
```

# Black Hat - Opp of Top Hat

```python
blackhat=cv.morphologyEx(img,cv.MORPH_BLACKHAT,kernal)
```

# Display blackhat

```python
cv.imshow('img',img)
cv.imshow('BLACK HAT',blackhat)
cv.waitKey(0)
cv.destroyAllWindows()
```
